```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

xls = 'form_answers.xlsx'
# Load each sheet to inspect its contents
df_simplified = pd.read_excel(xls, sheet_name='Simplified column names')
df_answers_performance = pd.read_excel(xls, sheet_name='From answers and performances')
df_form_responses = pd.read_excel(xls, sheet_name='Form Yanıtları 1')

# Display the first few rows of each sheet
df_simplified.head(), df_answers_performance.head(), df_form_responses.head()



# Selecting relevant columns related to time spent on different phases
time_columns = [
    "Time understanding definition",
    "Time spent on downloading data",
    "Time debugging apping",
    "Time for mapping",
    "Time for variant calling",
    "Time for filtering",
    "Time for analysis",
    "Time for visualization",
    "Time for report",
    "Time for presentation"
]


# Mapping time categories to numerical values
time_mapping = {
    "< 2 hours": 1,
    "2- 4 hours": 2,
    "4-8 hours": 3,
    "8-16 hours": 4,
    "> 16 hours": 5
}

# Convert time categories to numerical values
df_time_numeric = df_simplified.copy()
for col in time_columns:
    df_time_numeric[col] = df_time_numeric[col].map(time_mapping)

# Grouping by Group No and averaging time spent
df_time_grouped = df_time_numeric.groupby("Group No")[time_columns].mean()

# change font to times new roman
plt.rcParams["font.family"] = "Times New Roman"
# change figure text font size
plt.rcParams.update({'font.size': 12})

# rename the xticks by deleting "time for" part
df_time_grouped.columns = [col.replace("Time for ", "") for col in df_time_grouped.columns]
df_time_grouped.columns = [col.replace("Time ", "") for col in df_time_grouped.columns]
df_time_grouped.columns = [col.replace("spent on ", "") for col in df_time_grouped.columns]
df_time_grouped.columns = [col.replace(" apping", " mapping") for col in df_time_grouped.columns]

# capitalize each xtick text
df_time_grouped.columns = [col.title() for col in df_time_grouped.columns]


# Plotting the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_time_grouped, cmap="flare", linewidths=0.5, fmt=".1f")

plt.title("Average Time Spent on Different Phases by Group")
plt.xlabel("Project Phases")
plt.ylabel("Group No")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()

# Show the heatmap
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

xls = 'form_answers.xlsx'

# Selecting relevant columns related to time spent on different phases
time_columns = [
    "Time understanding definition",
    "Time installing cosap",
    "Time debugging download",
    "Time spent on downloading data",
    "Time debugging apping",
    "Time for mapping",
    "Time for variant calling",
    "Time for filtering",
    "Time for analysis",
    "Time for visualization",
    "Time for report",
    "Time for presentation"
]


# Load each sheet to inspect its contents
df_simplified = pd.read_excel(xls, sheet_name='Simplified column names')
df_answers_performance = pd.read_excel(xls, sheet_name='From answers and performances')
df_form_responses = pd.read_excel(xls, sheet_name='Form Yanıtları 1')

# Display the first few rows of each sheet
df_simplified.head(), df_answers_performance.head(), df_form_responses.head()
# Define time categories and corresponding numerical values
time_mapping = {
    "< 2 hours": 1,
    "2- 4 hours": 2,
    "4-8 hours": 3,
    "8-16 hours": 4,
    "> 16 hours": 5
}

# Convert time categories to numerical values
df_time_numeric = df_simplified.copy()
for col in time_columns:
    df_time_numeric[col] = df_time_numeric[col].map(time_mapping)

# Grouping by Group No and averaging time spent
df_time_grouped = df_time_numeric.groupby("Group No")[time_columns].mean()

# Define discrete color levels
levels = np.arange(0.5, 6, 1)  # Midpoints for discrete categories
cmap = sns.color_palette("muted", n_colors=len(levels) - 1)  # Discrete colormap

# change font to times new roman
plt.rcParams["font.family"] = "Times New Roman"
# change figure text font size
plt.rcParams.update({'font.size': 12})

# rename the xticks by deleting "time for" part
df_time_grouped.columns = [col.replace("Time for ", "") for col in df_time_grouped.columns]
df_time_grouped.columns = [col.replace("Time ", "") for col in df_time_grouped.columns]
df_time_grouped.columns = [col.replace("spent on ", "") for col in df_time_grouped.columns]
df_time_grouped.columns = [col.replace(" apping", " mapping") for col in df_time_grouped.columns]

# capitalize each xtick text
df_time_grouped.columns = [col.title() for col in df_time_grouped.columns]


# Plot heatmap with discrete colorbar
plt.figure(figsize=(8, 6))
ax = sns.heatmap(df_time_grouped, cmap=cmap, linewidths=0.5, cbar=True, vmin=1, vmax=5)

# Customize colorbar with discrete labels
cbar = ax.collections[0].colorbar
cbar.set_ticks(range(1, 6))  # Ensure ticks match discrete levels
cbar.set_ticklabels(["<2h", "2-4h", "4-8h", "8-16h", ">16h"])  # Custom labels

plt.title("Average Time Spent on Different Phases by Group")
plt.xlabel("Project Phases")
plt.ylabel("Group No")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()

plt.savefig("Form_Figures/time_heatmap.pdf", format="pdf", bbox_inches="tight")
plt.savefig("Form_Figures/time_heatmap.png", format="png", bbox_inches="tight")
# Show the heatmap
plt.show()
```

```python
# Load the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = 'form_answers.xlsx'
  # Update if needed
df_simplified = pd.read_excel(file_path, sheet_name="Simplified column names")

# Selecting categorical columns for the heatmap
categorical_columns = ["Method", "OS", "RAM", "Hard disk", "Most difficult part"]

# Keep only relevant columns and group data by Group No
df_heatmap = df_simplified.set_index("Group No")[categorical_columns]

# Replace long category names with shorter versions
df_heatmap.replace({
    "Cloning COSAP repository from Github": "Github",
    "COSAP image from Docker": "Docker"
}, inplace=True)

# Define a colormap for better contrast
cmap = sns.color_palette("dark", as_cmap=True)



# Create a heatmap with text labels inside the cells
plt.figure(figsize=(8, 6))
ax = sns.heatmap(
    df_heatmap.applymap(lambda x: hash(x) % 100),  # Convert categorical values to stable numerical values for coloring
    annot=df_heatmap,  # Show actual category labels inside cells
    fmt="",  # No numerical formatting
    cmap=cmap,  # Apply colormap
    linewidths=0.5,  # Add cell borders for better visibility
    cbar=False  # Remove colorbar since text is displayed
)

plt.title("Categorical Questions by Group")
plt.xlabel("Questions")
plt.ylabel("Group No")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()

plt.savefig("Form_Figures/heatmap_ram-os.pdf", format="pdf", bbox_inches="tight")
plt.savefig("Form_Figures/heatmap_ram-os.png", format="png", bbox_inches="tight")
# Show the heatmap
plt.show()
```
